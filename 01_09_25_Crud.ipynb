CREATE(Add new key-value pair)


```python
student={}#empty dictionary
student["name"]="Alice"#add new key name its value is alice
student["age"]=20 # add new key age, its value is 21
student["grade"]="A"
print(student)
```

```python
print(student["name"])
print(student.get("age"))
```

```python
print(student.get("grade"))
```

UPDATE

```python
student["name"]="bob"
student["age"]=25
student["grade"]="B"
print(student)
```

DELETE

```python
del student["grade"]# completely removes key and value
print(student)
```

```python
removed=student.pop("age")
#pop will delete the key, but value is present
#still cant access the value
print(student)
print(removed)
```

NESTED DICTIONARY

```python
#create a nested dictionary with student data
student={
    "Alice":{"Math":85, "Science":90,"English":78},
    "Cristiano":{"Math":82, "Science":91,"English":74},
    "Bob":{"Math":83, "Science":98,"English":75},
    "Mbappe":{"Math":87, "Science":91,"English":79},
    "Courtois":{"Math":98,"Science":93,"English":23}
}
student
```

```python
import pandas as pd #pandas are library for data manipulation
data = pd.DataFrame(student) # pass the dictionary to dataframe
data
```

```python
data=pd.DataFrame(student).T

data
```

```python
data["Average"]=data.mean(axis=1)
#axis=1 denotes row wise access to the dataFrame
data
```

```python
data["Math"]=data.mean(axis=0)
# axis=0 denotes colunm wiise to the dataframe
data
# the result is not a number(NaN) because there is  o variable to store the average
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
for subject in["Math","Science","English"]:
    plt.plot(data.index,data[subject],marker='o',label=subject)

plt.title("Student Average Performance")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.show()
```

```python

plt.figure(figsize=(8,5))
data[["Math","Science","English"]].plot(kind="box")

plt.title("Marks Distribution by Students")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.show()
```
